Simple demonstration of the ROC curve and its AUC for different binary probabilistic classifiers.

```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, roc_auc_score

from src import plot_roc_curve
```

```python
# Generate a synthetic dataset which we'll use to evaluate the classifiers.
x, y = make_classification(
    n_samples=1000,
    n_features=20,
    n_informative=2,
    n_redundant=10,
    random_state=42,
)
```

# Logistic Regression

```python
clf = LogisticRegression().fit(x, y)
y_pred_prob = clf.predict_proba(x)[:, 1]
fpr, tpr, thresholds = roc_curve(y, y_pred_prob)
auc = roc_auc_score(y, y_pred_prob)
print(f"{auc=}")
plot_roc_curve(fpr, tpr, thresholds).draw()
```

# Random Classifier

```python
rng_pred_prob = np.random.rand(x.shape[0])
fpr, tpr, thresholds = roc_curve(y, rng_pred_prob)
auc = roc_auc_score(y, rng_pred_prob)
print(f"{auc=}")
plot_roc_curve(fpr, tpr, thresholds).draw()
```

# Perfect Classifier

```python
perf_pred_prob = y.astype(float)
fpr, tpr, thresholds = roc_curve(y, perf_pred_prob)
auc = roc_auc_score(y, perf_pred_prob)
print(f"{auc=}")
plot_roc_curve(fpr, tpr, thresholds).draw()
```
